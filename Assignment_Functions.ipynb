Q1. What is the difference between a function and a method in Python?

**Ans** - In Python, both functions and methods are used to perform specific tasks or operations, but they are used in slightly different ways and belong to different contexts.

**a. Function**

A function is a block of reusable code that performs a specific task. It is defined using the def keyword and can be called independently by its name. Functions can take inputs (called parameters) and return outputs.

Example of a function:

def add_numbers(a, b):

    return a + b

result = add_numbers(5, 3)

print(result)   # Output: 8

Here, add_numbers() is a function that takes two numbers and returns their sum.

**b. Method**

A method is similar to a function, but it is associated with an object and is called on that object. Methods are defined inside classes and act on the data (attributes) of that class. Methods are always accessed using dot (.) notation.

Example of a method:

text = "hello world"

print(text.upper())  # Output: HELLO WORLD

Here, upper() is a method that belongs to the str class and is used to convert the string to uppercase. It is called on the string object text.

**Key Differences:**

**Feature**------|----------**Function**----------|-----------------------------**Method**

a. Called on -------|-------Can be called independently-----|----	Called on an object

b. Defined using------|-----def keyword---------------|-----Defined inside a class

c. Belongs to---------|-------Global or module level----|	Object of a class

d. Example-------------|---------	len([1, 2, 3])-------------------|-------"abc".upper()

**Application:**

a. Use a function when you need to perform an operation globally or on input values.

b. Use a method when you want to perform operations that depend on the objectâ€™s data (like a string or list).

**What is an Iterable?**

An iterable is any Python object you can loop over using a for loop. Examples include:

a. Lists

b. Tuples

e. Dictionaries

**Key Point: An iterable doesnâ€™t keep track of its state during iteration.**

Example of an Iterable:

    print(item)

**What is an Iterator?**

An iterator is an object that represents a stream of data; it remembers where it is during iteration and can be used to fetch the next value using the next() function.

To get an iterator from an iterable, we use the built-in iter() function.

Example of an Iterator:

my_list = [10, 20, 30]

print(next(it))  # Output: 10

print(next(it))  # Output: 20

print(next(it))  # Output: 30

If you call next(it) again, it will raise a StopIteration error because the iterator has no more items.

Q6. Explain the concept of generators in Python and how they are defined.

**Ans**-In Python, a generator is a special type of function that returns an iterator and allows us to generate values on the fly instead of storing them all in memory at once. This makes generators memory-efficient and ideal for working with large datasets.

**What Makes Generators Special?**

a. Unlike normal functions that return a single value using return, generators use the yield keyword to return a sequence of values, one at a time.

b. The function remembers its state between calls automatically.

c. Generators are lazy, meaning they generate values only when asked, which saves memory.

**How to Define a Generator:**

You define a generator just like a normal function, but use yield instead of return.

def count_up_to(n):

    count = 1

    while count <= n:

        yield count

        count += 1

This generator will yield numbers from 1 up to n.

**How to Use a Generator:**

You can loop through a generator using a for loop, or use next() to get each value.

counter = count_up_to(3)

print(next(counter))  # Output: 1

print(next(counter))  # Output: 2

print(next(counter))  # Output: 3

If you call next(counter) again, it will raise StopIteration.

**Generator vs Normal Function**

**Feature**----------|---------**Normal Function**--------|--------**Generator Function**

Returns all values at once---------|---- Yes---------------|------------No

Uses return-------------------|------------Yes------------|-------------No

Uses yield ------------------|--------------No------------|------------Yes

Memory efficient---------|--------Can use more memory-----|-----Very memory efficient

Real-Life Example: Reading Large Files

Imagine you're reading a 1 GB CSV file. Using a generator can save memory:

def read_lines(file_path):

    with open(file_path, 'r') as file:

        for line in file:

            yield line

Instead of reading the entire file into memory, the generator reads it line-by-line, making your program faster and more efficient.

**Summary**

a. Generators use yield to return values one at a time.

b. They are memory efficient and remember where they left off.

c. Ideal for looping through large data without loading everything at once.

Q7. What are the advantages of using generators over regular functions?

**Ans** - Generators in Python provide a powerful way to iterate over large data without loading everything into memory. They are defined using the yield keyword and offer several advantages over regular functions, especially in data analysis where performance and memory usage matter.

**Key Advantages of Using Generators:**

1. Memory Efficiency- Generators donâ€™t store all values in memory. They generate each value on demand. Ideal for large datasets, such as reading huge files or generating large number sequences.

Example:

def big_numbers():

    for i in range(1, 1000000):

        yield i

If you used range() or stored all values in a list, it would take a lot of RAM. But the above generator gives values one by one.

2. Lazy Evaluation (On-Demand Execution) -The function pauses at each yield and resumes from the same point next time. This reduces processing time when not all data is needed immediately.

Example:

def infinite_counter():

    i = 1

    while True:

        yield i

        i += 1

This infinite generator doesnâ€™t crash your program. It keeps yielding values one at a time only when asked.

3. Improves Performance in Loops- Since generators donâ€™t pre-calculate values, loops that process data run faster when not all data is needed up front.

4. Cleaner and More Readable Code - Generators make the code simpler and more Pythonic for certain tasks like file reading, data streaming, or filtering data.

Example: Reading a large file:

def read_lines(file_path):

    with open(file_path, 'r') as file:

        for line in file:

            yield line

Efficiently reads one line at a time instead of loading the entire file into memory.

**Comparison Table:**

**Featur**-----------|------------Regular Function-----------|--------Generator Function

Returns--------------|----------All values at once (return)----|------One value at a time (yield)

Memory usage----------|-----------High (stores all values)-------|------Low (no storage needed)

Execution-------------------|------------Immediate-----------|-------On demand (lazy)

Use case---------------------|----------Small datasets--------|---------Large / infinite datasets

**Summary:**

a.Generators are memory-friendly, efficient, and ideal for processing large data.

Q4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example: Check and print all prime numbers from 1 to 200
print("Prime numbers from 1 to 200 are:")
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_val = self.a + self.b
            self.a, self.b = self.b, next_val
            self.count += 1
            return next_val

# Example usage
fib = FibonacciIterator(10)
for num in fib:
    print(num, end=" ")
```

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i

# Example usage
print("Powers of 2 up to exponent 5:")
for power in powers_of_two(5):
    print(power, end=" ")
```

Q10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(text):
    vowels = 'aeiouAEIOU'
    # Use filter() with lambda to exclude vowels
    result = ''.join(filter(lambda ch: ch not in vowels, text))
    return result

# Example usage
input_string = "Data Analysis with Python"
output_string = remove_vowels(input_string)
print("String without vowels:", output_string)
```

Q11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

98762	Programming Python, Mark lutz	5	56.8

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- â‚¬ if the value of the
order is smaller than 100,00 â‚¬.

Write a Python program using lambda and map

```python
# Order data as a list of lists
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Baryy", 3, 32.95],
    [88112, "Einfuhrung in Python3, Bernd Klein", 3, 24.99]
]

def order_summary(orders):
    result = []
    for order in orders:
        order_no = order[0]
        total_price = order[2] * order[3]
        if total_price < 100:
            total_price += 10
        result.append((order_no, round(total_price, 2)))
    return result

# Example usage
summary = order_summary(orders)
print(summary)
```
